```python
from matplotlib import pyplot as plt
import matplotlib.patches as patches

def plt_person(img, new_box, save_path):
    fig = plt.figure()
    plt.axis('off')
    img = np.array(img)
    rect = patches.Rectangle((int(new_box[0]), int(new_box[1])), int(new_box[2]-new_box[0]), int(new_box[3]-new_box[1]), linewidth=3, edgecolor=(0,1,0), facecolor='none')
    plt.imshow(img)
    ax = plt.gca()
    ax.add_patch(rect)
    #plt.show()
    print(save_path)
    plt.savefig(save_path)
    
```

```python
def get_vqa_answer(image, question, generator, models, task, image_transform, use_cuda=True, use_fp16=False):
    # Construct input sample & preprocess for GPU if cuda available
    sample = construct_sample_vqa(task, image_transform, image, question)
    sample = utils.move_to_cuda(sample) if use_cuda else sample
    sample = utils.apply_to_sample(apply_half, sample) if use_fp16 else sample
    # Run eval step for open-domain VQA
    with torch.no_grad():
        result, scores = zero_shot_step(task, generator, models, sample)
    return result, scores
```
