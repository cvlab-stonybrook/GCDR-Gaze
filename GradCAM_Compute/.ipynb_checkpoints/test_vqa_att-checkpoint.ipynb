# Pre-process

## Process Image

```python
# Image transform
from torchvision import transforms
mean = [0.5, 0.5, 0.5]
std = [0.5, 0.5, 0.5]

patch_resize_transform = transforms.Compose([
    lambda image: image.convert("RGB"),
    transforms.Resize((cfg.task.patch_image_size, cfg.task.patch_image_size), interpolation=Image.BICUBIC),
    transforms.ToTensor(),
    transforms.Normalize(mean=mean, std=std),
])

# Text preprocess
bos_item = torch.LongTensor([task.src_dict.bos()])
eos_item = torch.LongTensor([task.src_dict.eos()])
pad_idx = task.src_dict.pad()
```

## Process Text

```python

# Normalize the question
def pre_question(question, max_ques_words):
    question = question.lower().lstrip(",.!?*#:;~").replace('-', ' ').replace('/', ' ')
    question = re.sub(
        r"\s{2,}",
        ' ',
        question,
    )
    question = question.rstrip('\n')
    question = question.strip(' ')
    # truncate question
    question_words = question.split(' ')
    if len(question_words) > max_ques_words:
        question = ' '.join(question_words[:max_ques_words])
    return question

def encode_text(text, length=None, append_bos=False, append_eos=False):
    s = task.tgt_dict.encode_line(
        line=task.bpe.encode(text),
        add_if_not_exist=False,
        append_eos=False
    ).long()
    if length is not None:
        s = s[:length]
    if append_bos:
        s = torch.cat([bos_item, s])
    if append_eos:
        s = torch.cat([s, eos_item])
    return s

# Construct input for open-domain VQA task
def construct_sample(image: Image, question: str):
    patch_image = patch_resize_transform(image).unsqueeze(0)
    patch_mask = torch.tensor([True])
    question = pre_question(question, task.cfg.max_src_length)
    question = question + '?' if not question.endswith('?') else question
    #Pdb().set_trace()
    src_text = encode_text(' {}'.format(question), append_bos=True, append_eos=True).unsqueeze(0)

    src_length = torch.LongTensor([s.ne(pad_idx).long().sum() for s in src_text])
    ref_dict = np.array([{'yes': 1.0}]) # just placeholder
    sample = {
        "id":np.array(['42']),
        "net_input": {
            "src_tokens": src_text,
            "src_lengths": src_length,
            "patch_images": patch_image,
            "patch_masks": patch_mask,
        },
        "ref_dict": ref_dict,
    }
    return sample
  
# Function to turn FP32 to FP16
def apply_half(t):
    if t.dtype is torch.float32:
        return t.to(dtype=torch.half)
    return t
```

# Inference

```python
def decode_fn(x, tgt_dict, bpe, generator, tokenizer=None):
    x = tgt_dict.string(x.int().cpu(), extra_symbols_to_ignore={generator.bos, generator.eos})
    if bpe is not None:
        x = bpe.decode(x)
    if tokenizer is not None:
        x = tokenizer.decode(x)
    return x

def get_str_from_tokens(result, task, generator):
    info = []
    for idx, beam_out in enumerate(result):
        tokens = beam_out["tokens"]
        detok_hypo_str = decode_fn(tokens, task.tgt_dict, task.bpe, generator)
        info.append({"tokens":tokens, "str": detok_hypo_str, "score":beam_out["score"]})
    return info

def get_vqa_result(image, question, model, decoder_att_layer=11, use_cuda=True, use_fp16=False, get_encoder_att=False):
    sample = construct_sample(image, question)
    sample = utils.move_to_cuda(sample) if use_cuda else sample
    sample = utils.apply_to_sample(apply_half, sample) if use_fp16 else sample
    # Run eval step for open-domain VQA
    with torch.no_grad():
        result = model(sample["net_input"], get_encoder_att=get_encoder_att, decoder_alignment_layer=decoder_att_layer)
    encoder_att = []
    if get_encoder_att:
        encoder_att = result[1][:,0,:,:].detach().cpu().numpy()  # shape (num_layer * seq_len * seq_len)
        result = result[0]
    result = result[0]  # first batch sample
        
    return sample, result, encoder_att
    
```

```python
def plot_encoder_att_weights(image, encoder_att_weights,step, layer=-1,plot_img_size=(224,224)):
    plt.figure()
    plt.axis('off')
    width, height = image.size
    image.convert('RGB')
    image = image.resize(plot_img_size, resample=Image.BILINEAR)
    image = np.asarray(image)
    #encoder_att_weights = encoder_att_weights[layer,900+step,:900].reshape(30,30)
    encoder_att_weights = encoder_att_weights[layer,:900,900+step].reshape(30,30)
    resized_encoder_att = resize(encoder_att_weights, plot_img_size)
    resized_encoder_att = (resized_encoder_att - resized_encoder_att.min()) / (resized_encoder_att.max() - resized_encoder_att.min())
    plt.imshow(image)
    plt.imshow(resized_encoder_att, cmap='jet', vmin=0.0, vmax=1.0, alpha=0.3)
    plt.show()
    plt.close()
```

```python
def plot_decoder_att_weights(image, result, step, plot_img_size=(224,224)):
    plt.figure()
    plt.axis('off')
    width, height = image.size
    image.convert('RGB')
    image = image.resize(plot_img_size, resample=Image.BILINEAR)
    image = np.asarray(image)
    decoder_att_weights = result["attention"].cpu().numpy()
    decoder_att_weights = decoder_att_weights[:900,step].reshape(30,30)
    resized_decoder_att = resize(decoder_att_weights, plot_img_size)
    resized_decoder_att = (resized_decoder_att - resized_decoder_att.min()) / (resized_decoder_att.max() - resized_decoder_att.min())
    plt.imshow(image)
    plt.imshow(resized_decoder_att, cmap='jet', vmin=0.0, vmax=1.0, alpha=0.3)
    plt.show()
    plt.close()
    return resized_decoder_att
```
